```python
num_cols = [
'Login_Frequency',
'Session_Duration_Avg',
'Pages_Per_Session',
'Total_Purchases',
'Average_Order_Value',
'Cart_Abandonment_Rate',
'Customer_Service_Calls',
'Lifetime_Value'
]

for col in num_cols:
    print("\n",col)
    print(df.groupby("Churned")[col].mean())
```

```python
num_cols = df.select_dtypes(
    include=['float64','int64']
).columns

for col in num_cols:

    if df[col].isnull().sum()>0:

        median = df[col].median()

        df[col].fillna(
            median,
            inplace=True
        )
```

```python
cat_cols = df.select_dtypes(
    include='object'
).columns

for col in cat_cols:

    mode = df[col].mode()[0]

    df[col].fillna(
        mode,
        inplace=True
    )
```
